```python
def to_detect_outliers(X,features):
    final_outlier_index = []
    for col in features:
        q1 = np.percentile(X[col],25)
        q3 = np.percentile(X[col],75)
        IQR = (q3 - q1) * 1.5
        lower_limit = q1 - IQR
        upper_limit = q3 + IQR
        outlier_index = X[col][(X[col]<lower_limit)|(X[col]>upper_limit)].index.to_list()
        final_outlier_index.extend(outlier_index)
    out_index = list(set(final_outlier_index))
    out_index.sort()
    print(out_index)
    print("----------------------------------------------------------------------")
    print(f'There are {len(out_index)} records which fall under outliers' )
    print("----------------------------------------------------------------------")
    print(f'the percentage of outliers in the dataset is : {round((len(out_index)/len(df)*100),2)}% ')
        
        
    
```

#### Capping Outliers

```python
def capping_outliers(df,feature):
    for col in feature:
        q1 = np.percentile(df[col],25)
        q3 = np.percentile(df[col],75)
        IQR = (q3-q1) * 1.5
        upper_bond = q3 + IQR
        lower_bond = q1 - IQR
        df[col] = np.where(df[col]>upper_bond,upper_bond,np.where(df[col]<lower_bond,lower_bond,df[col]))
        #name_of_col      if is true , replace it with upper bond , else (replace is lower bond if both 
        # condition are failed then keep remaining data as it  )
        
```

##### Xgboost Hyperparameter Tuning

XGBClassifier()
1. max_depth: The maximum depth of each tree in the ensemble. Increasing this value will make the model more complex and may 
   lead to overfitting.

2. learning_rate: The learning rate controls the contribution of each tree in the ensemble. A smaller learning rate will lead to
   slower convergence but a more accurate model.

3. n_estimators: The number of trees in the ensemble. A larger number of trees will make the model more robust but also increase
   the training time.

4. subsample: The fraction of observations to be sampled for each tree. Smaller values will make the model more robust but also increase the training time.

5. colsample_bytree: The fraction of features to be used for each tree. Smaller values will make the model more robust but also increase the training time.

6. min_child_weight: Minimum sum of weights of all observations required in a child. A smaller value will lead to deeper trees and more complex model.

7. reg_alpha: L1 regularization term on weights. Increasing this value will make model more conservative.

8. reg_lambda: L2 regularization term on weights. Increasing this value will make model more conservative.
